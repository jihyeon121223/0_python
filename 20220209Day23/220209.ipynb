```python
import random
names = '가나다라마바사아자차카타파하'
with open('student_info.csv','w')as f:
    for i in range(1000):
        name = random.choice(names)+random.choice(names)+random.choice(names)
        height=random.randrange(150,190) #150 이상 190 미만의 난수
        weight=random.randrange(40,90)
        f.write(f'{name},{height},{weight}\n')
```

```python
with open('student_info.csv','r') as f:
    for line in f:
        (name,height,weight)=line.strip().split(',')
        bmi=int(weight)/(int(height)**2)
        result=''
        if 25 <= bmi:
            result='과체중'
        elif 18.5<=bmi:
            result='정상'
        else:
            result='저체중'
        print('\n'.join([
            '이름:{}',
            '몸무게:{}',
            '키:{}',
            'bmi:{}',
            '결과:{}',
        ]).format(name,weight,height,bmi,result))
```

```python
odds=[]
for number in numbers:
    if number %2 !=0:
        odds.append(number)
print(odds)
odds=[number for number in numbers if number %2 !=0] #방법1
print(odds) #방법2
print(list(filter(lambda x: x%2 !=0,numbers))) #방법3
```

```python
odds=[]
for number in numbers:
    if number %2 !=0:
        odds.append(number)
print(odds)
for odd in odds:
    if odd >= 3 and odd < 7:
        print(odds)
odds=[number for number in numbers if number %2 !=0]
print(odds)
print(list(filter(lambda x: x%2 !=0,numbers)))
```

```python
new_obj=[]
for number in numbers:
    if number>=3 and number<7:
        new_obj.append(number)
print(new_obj)
new_obj=[number for number in numbers if number >=3 and number <7]
print(new_obj)
print(list(filter(lambda x: x>=3 and x<7,numbers)))
```
